# Saving Image for NFIQ2.0
- nfiq 2.0 require 500 ppi image
- .png extentions

```
IMAGE_PATH = '/content/drive/MyDrive/test_noise_reshaped/'
IMAGE_TARGET_PATH = '/content/drive/MyDrive/data_samples/synth_data_500dpi/input_test_reshape/'
```

```
# Get image paths from the directory
image_path_list = [f for f in glob.glob(IMAGE_PATH + '*.png')]
image_path_list.sort()


print('image path lenght:', len(image_path_list))

image_dataframe = {
  'image': image_path_list,
}

image_df = pd.DataFrame(data = image_dataframe)
print(image_df)
```

```
def extract_image_name(path_list):
  extracted_list = []
  for path in path_list:
    file_name = os.path.basename(path)
    extracted_list.append(file_name)
  return extracted_list
```

```
index = extract_image_name(image_df.image)
```

```
index
```

```
def set_dpi_image_batch(in_paths, index, out_path, dpi=500):

  for i,path in enumerate(in_paths):
    img = Image.open(path)
    name = index[i]
    if (i % 250 == 0):
      print("progress: ", i)
    img.save(out_path + name, dpi=(dpi, dpi))
```

```
set_dpi_image_batch(image_df.image[:], index[:], IMAGE_TARGET_PATH, 500)
```
